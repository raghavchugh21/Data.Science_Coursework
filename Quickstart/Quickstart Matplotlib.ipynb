```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
```

```python
y = x**2 +3
print(y)
```

```python
plt.plot(x,y,label = "Stock Price 2020")
plt.plot(x,x**2,label = "Stock Price 2019")
plt.title("Stock Market")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```
